데이터는 환자에 건강정보에 따라 outcome을 예측하는 모델입니다.
이번에는 각 코드에 관해서 분석하는 텍스트가 추가됩니다.

```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score



# -----------------------------

# 1) 데이터 준비

# -----------------------------

df = pd.read_csv("/content/drive/MyDrive/diabetes.csv")
df
```

아래 과정은 데이터에서 입력데이터가될 부분을 X로, 출력데이터 역할을할 데이터를 나누는 과정입니다.

```python
X = df.drop(columns=["Outcome"])

y = df["Outcome"]
```

X와 y데이터를 확인하는 표

```python
X
```

```python
y
```

아래는 X와 y데이터를 각각 훈련과 테스트 데이터로 나누는 과정입니다.
test_size는 테스트 데이터로 넘어갈 비율을 의미합니다.
stratify=y는 y를 구성하는 데이터 비율을 훈련과 테스트 데이터로 나눌때 그 비율을 유지시킬지의 여부입니다.

```python
X_train, X_test, y_train, y_test = train_test_split(

    X, y, test_size=0.2, stratify=y, random_state=42

)
```

이번 모델에서 사용할 3가지 분류모델입니다.

```python
dt = DecisionTreeClassifier(random_state=42)

rf = RandomForestClassifier(n_estimators=200, random_state=42)

lr = LogisticRegression(max_iter=500)
```

각 분류모델에 훈련 데이터를 넣고 훈련시키는 과정입니다.

```python
dt.fit(X_train, y_train)

rf.fit(X_train, y_train)

lr.fit(X_train, y_train)
```

predict는 훈련된 모델들에 X_test를 대입 후 이를 실제 데이터 y_test와 비교하는 과정입니다. 정확도가 높을 수록 숫자는 1에 가까워집니다.

```python
dt_acc = accuracy_score(y_test, dt.predict(X_test))

rf_acc = accuracy_score(y_test, rf.predict(X_test))

lr_acc = accuracy_score(y_test, lr.predict(X_test))
```

각 모델들의 정확도를 출력한 결과입니다.

```python
print("=== Test Accuracy ===")

print(f"Decision Tree : {dt_acc:.4f}")

print(f"Random Forest : {rf_acc:.4f}")

print(f"Logistic Reg. : {lr_acc:.4f}")
```
