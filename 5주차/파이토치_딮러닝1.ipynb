```python
#데이터 불러오기
import matplotlib.pyplot as plt
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split

digits = datasets.load_digits()
plt.imshow(digits.images[0], cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

```python
#이미지를 1차원으로 벡터화
n_samples = len(digits.images)

data = digits.images.reshape((n_samples, -1))
```

```python
#위 데이터를 테스트, 훈련 데이터로 나누는 과정
X_train, X_test, y_train, y_test = train_test_split(data, digits.target, test_size=0.2)
```

```python
#KNN 모델 생성 및 학습
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
#KNN 모델의 데이터 예측, 및 결과 출력
knn_pred = knn.predict(X_test)

knn_scores = metrics.accuracy_score(y_test, knn_pred)
print(knn_scores)
```

```python
#이미지를 출력하기 위해서 다시 2차원으로 올리는 과정
plt.imshow(X_test[0].reshape(8, 8), cmap=plt.cm.gray_r, interpolation='nearest')
```

```python
#DT 모델 생성 및 학습
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
#DT 모델의 데이터 예측, 및 결과 출력
dt_pred = dt.predict(X_test)

dt_scores = metrics.accuracy_score(y_test, dt_pred)
print(dt_scores)
```

```python
#LR 모델 생성 및 학습
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
#LR 모델의 데이터 예측, 및 결과 출력
lr_pred = lr.predict(X_test)

lr_scores = metrics.accuracy_score(y_test, lr_pred)
print(lr_scores)
```

```python
#RF 모델 생성 및 학습
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
#RF 모델의 데이터 예측, 및 결과 출력
rf_pred = rf.predict(X_test)

rf_scores = metrics.accuracy_score(y_test, rf_pred)
print(rf_scores)
```

```python
#최종적인 각 모델별 예측 정확도
print("knn : ", knn_scores)
print("dt : ", dt_scores)
print("lr : ", lr_scores)
print("rf : ", rf_scores)
```
